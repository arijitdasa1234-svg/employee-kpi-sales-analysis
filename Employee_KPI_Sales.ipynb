```python
df_csv["Sales_Difference"] = df_csv["Sales_Achieved"] - df_csv["Sales_Target"]

plt.figure(figsize=(10,5))

plt.hist(df_csv["Sales_Difference"],
         bins=30,
         edgecolor="black")

plt.title("Sales Difference (Achieved - Target)")
plt.xlabel("Difference (INR)")
plt.ylabel("Number of Employees")

plt.show()
```

```python
print("describe----->\n",df_csv['Sales_Target'].describe(),"\n")
print("describe----->\n",df_csv['Sales_Achieved'].describe(),"\n")

import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

# Maximum value for both axes
max_value = max(df_csv["Sales_Target"].max(),
                df_csv["Sales_Achieved"].max())

# Draw only the reference line
plt.plot([0, max_value],
         [0, max_value],
         color="red",
         linestyle="--",
         linewidth=2,
         label="Target = Achieved")

plt.title("Sales Target = Sales Achieved")
plt.xlabel("Sales Target (INR)")
plt.ylabel("Sales Achieved (INR)")
plt.legend()
plt.grid(True)

plt.show()

print("\n Red dashed line: Employees who exactly met their sales target.\n")
print("\nPoints above the line: Employees who exceeded their target\n")
print("\nPoints below the line: Employees who did not meet their target.\n")
```

```python
def kpi_category(kpi):
    if kpi >= 90:
        return "Best"
    elif kpi >= 80:
        return "Good"
    elif kpi >= 70:
        return "Average"
    elif kpi >= 60:
        return "Needs Improvement"
    else:
        return "Poor"

# Create a new column
df_csv["KPI_Category"] = df_csv["Monthly_KPI"].apply(kpi_category)

# Display Employee ID, KPI, and Category
print(df_csv[["Employee_ID", "Monthly_KPI", "KPI_Category"]])
```

```python
kpi_order = ["Poor", "Needs Improvement", "Average", "Good", "Best"]

print(
    df_csv["KPI_Category"]
    .value_counts()
    .reindex(kpi_order)
)
```

```python
import matplotlib.pyplot as plt

order = ["Poor", "Needs Improvement", "Average", "Good", "Best"]

df_csv["KPI_Category"].value_counts().reindex(order).plot(
    kind="bar",
    edgecolor="black"
)

plt.title("Employee KPI Categories")
plt.xlabel("KPI Category")
plt.ylabel("Number of Employees")
plt.grid(axis="y", alpha=0.3)

plt.show()
```
